```
def list_relu(L):
  new_list=L.copy()
  for i in range(len(L)):
    if L[i]<0:
      new_list[i]=0
    else:
      new_list[i]=L[i] 
  return new_list
```

**Problem 7**

```
def matrix_sum(M):
  n_rows= len(M)
  n_cols= len(M[0,:])
  sum=0
  for i in range(n_rows):
    for j in range(n_cols):
      sum += M[i,j]
  return sum
```
