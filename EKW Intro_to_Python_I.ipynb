```
def swap(c):
  new_list=c.copy() # This creates a copy of the list c, which is called new_list.  Your function should work with this list instead of the list c which is passed to the function.
  # Put your code here.
  temp_var = new_list[0] # save value of the first entry as a temporary variable
  new_list[0] = new_list[-1] # set the first entry of the list to the value of the last entry of the list
  new_list[-1] = temp_var # set the last entry to the value of the temporary variable storing the original first entry
  return new_list
```

**Problem 8**

```
def trunc_max(x,y):
  # Put your code here.
  if x >= y:
    return_int = x
  if y > x:
    return_int = y
  if x <= 0 and y <= 0:
    return_int = 0
  return return_int
```
