**Problem 1**

```
def augment(A,b): 
  Ab = np.column_stack((A,b))
  return Ab
```

**Problem 2**

```
def first_column_zeros(A):
  B=np.copy(A)
  for i in range(1,len(A)):
    if (i != 0):
      B[i, :] = A[i, :] - ((A[i, 0]/A[0, 0])*A[0])
  return B
```

**Problem 3**

```
def row_echelon(A,b): 
  B = augment(A,b)
  for i in range(len(B)):
    B[i:, i:] = first_column_zeros(B[i:, i:])
  return B
```

**Problem 4**

```
def LU_decomposition(A):
  m, n = np.shape(A)
  U=np.copy(A)
  L=np.identity(len(A))
  for j in range(0, n):
    for i in range(j+1, m):
      L[i, j] = (U[i, j]/U[j, j])
      U[i, :] = U[i, :] - (L[i, j]*U[j, :])
  return L,U
```

**Problem 5**

```
def forward_substitution(L,b):
  n = len(b)
  y = np.zeros(n)
  for i in range(0, n):
    y[i] = ((b[i] - np.dot(y, L[i, :]))/L[i, i])
  return y
```

**Problem 6**

```
def back_substitution(U,c):
  n = len(c)
  z = np.zeros(n)
  for i in reversed(range(0, n)):
    z[i] = ((c[i] - np.dot(z, U[i, :]))/U[i, i])
  return z
```

**Problem 7**

```
def LU_solver(A,b): 
  L,U = LU_decomposition(A)
  y = forward_substitution(L,b)
  return back_substitution(U,y)
```
